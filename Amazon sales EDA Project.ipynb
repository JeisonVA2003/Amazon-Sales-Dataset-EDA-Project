Changing Data Types of Columns `discounted_price` , `actual_price` , `discount_percentage` , `rating` ,  `rating_count`    from object to float

```python
df['discounted_price'] = df['discounted_price'].str.replace("â¹",'')
df['discounted_price'] = df['discounted_price'].str.replace(",",'')
df['discounted_price'] = df['discounted_price'].astype('float64')

df['actual_price'] = df['actual_price'].str.replace("â¹",'')
df['actual_price'] = df['actual_price'].str.replace(",",'')
df['actual_price'] = df['actual_price'].astype('float64')
```

```python
df['discount_percentage'] = df['discount_percentage'].str.replace('%','').astype('float64')

df['discount_percentage'] = df['discount_percentage'] / 100
```

```python
df['rating'].value_counts()
```

We found an unusual string in rating column `|`

So we are going to check the strange row

```python
df.query('rating == "|"')
```

We identify that this strange string is like a missing value and we have to search the product info in Amazon

I got this product rating on Amazon by searching the provided product_id on their official website (amazon.in)
The rating is 3.9. So, I am going to give the item rating a 3.9 as well.

```python
df['rating'] = df['rating'].str.replace('|', '3.9').astype('float64')
```

```python
df['rating_count'] = df['rating_count'].str.replace(',', '').astype('float64')
```

We will use histograms and box plots to understand the distribution, central tendency, and spread of the numerical data

#### 1. Distribution of Prices (actual_price)

```python
# Set the visual style for the plots
# sns.set_style("whitegrid")

# # Histogram for the actual price
plt.figure(figsize=(10, 6))
sns.histplot(df['actual_price'], kde=True, bins=30, color='skyblue')
plt.title('Distribution of Actual Prices', fontsize=16, weight='bold')
plt.xlabel('Actual Price ($)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

#### 2. Distribution of Ratings (rating)
This will show us whether customers tend to leave high or low ratings.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['rating'], kde=True, bins=20, color='coral')
plt.title('Distribution of Product Ratings', fontsize=16, weight='bold')
plt.xlabel('Rating (1.0 to 5.0)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

#### 3. Distribution of Rating Counts (rating_count)
This plot is crucial for understanding popularity. Do a few products have many reviews, or is it evenly distributed?

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['rating_count'], kde=True, bins=30, color='lightgreen')
plt.title('Distribution of Rating Counts per Product', fontsize=16, weight='bold')
plt.xlabel('Number of Ratings', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

#### 4. Distribution of Product Categories (category)
This will allow us to identify which categories have the most products in the dataset.

```python
top_categories = df['category'].value_counts().head(15)

plt.figure(figsize=(12, 8))
sns.barplot(x=top_categories.values, y=top_categories.index, palette='viridis', orient='h')
plt.title('Top 15 Categories by Product Count', fontsize=16, weight='bold')
plt.xlabel('Number of Products', fontsize=12)
plt.ylabel('Category', fontsize=12)
plt.show()
```

## Correlation Analysis 

```python
df_1 = df.select_dtypes(include=['float64', 'int64'])

correlation_matrix = df_1.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(
    correlation_matrix, 
    annot=True,          
    cmap='coolwarm',     
    fmt='.2f',          
    linewidths=.5       
)

plt.title('Correlation Matrix of Variables (Pearson)', fontsize=16, weight='bold')
plt.show()
```

```python
spearman_correlation_matrix = df_1.corr(method="spearman")
sns.heatmap(spearman_correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix (Spearman)")
plt.show()
```

```python
# Calculate correlation coefficient between product price and sales
correlation_coefficient = np.corrcoef(df['actual_price'], df['rating'])[0, 1]

# Print correlation coefficient
print(correlation_coefficient)
```

## Grouping and Aggregation

```python
mean_ratings = df.groupby('category')['rating'].mean()

# Sort in descending order and convert to a DataFrame for better printing
sorted_ratings_df = mean_ratings.sort_values(ascending=False).reset_index()
sorted_ratings_df.rename(columns={'rating': 'average_rating'}, inplace=True)

# Print the top 10 highest-rated categories
print("--- Top 10 Highest-Rated Categories ---")
print(sorted_ratings_df.head(10))
```

```python
top_15_categories = sorted_ratings_df.head(15)

plt.figure(figsize=(12, 8))
sns.barplot(
    x='average_rating', 
    y='category', 
    data=top_15_categories, 
    palette='viridis',
    orient='h'
)

# Add titles and labels for clarity
plt.title('Average Rating by Product Category (Top 15)', fontsize=16, weight='bold')
plt.xlabel('Average Rating (out of 5)', fontsize=12)
plt.ylabel('Category', fontsize=12)
plt.xlim(4, 5) # Optional: Zoom in on the high ratings if they are all close

plt.show()
```

```python
top_10 = sorted_ratings_df.head(10)
bottom_10 = sorted_ratings_df.tail(10).sort_values('average_rating', ascending=True)

# Create two subplots side-by-side
fig, axes = plt.subplots(1, 2, figsize=(18, 8))
fig.suptitle('Highest and Lowest Rated Product Categories', fontsize=20, weight='bold')

# --- Plot for Top 10 ---
sns.barplot(ax=axes[0], x='average_rating', y='category', data=top_10, palette='Greens_r')
axes[0].set_title('Top 10 Highest-Rated', fontsize=16)
axes[0].set_xlabel('Average Rating')
axes[0].set_ylabel('Category')

# --- Plot for Bottom 10 ---
sns.barplot(ax=axes[1], x='average_rating', y='category', data=bottom_10, palette='Reds_r')
axes[1].set_title('Top 10 Lowest-Rated', fontsize=16)
axes[1].set_xlabel('Average Rating')
axes[1].set_ylabel('') # Hide y-label for cleaner look

plt.tight_layout(rect=[0, 0, 1, 0.96]) # Adjust layout to make space for suptitle
plt.show()
```

```python
# 1. Group by CATEGORY and calculate the standard deviation of price
price_std_by_category = df.groupby('category')['actual_price'].std().sort_values(ascending=False)

# 2. Visualize the top 15 categories with the highest price variation
plt.figure(figsize=(12, 8))
sns.barplot(
    x=price_std_by_category.head(15).values,
    y=price_std_by_category.head(15).index,
    palette='plasma',
    orient='h'
)
plt.title('Top 15 Categories by Price Variation (Std. Deviation)', fontsize=16, weight='bold')
plt.xlabel('Standard Deviation of Price ($)')
plt.ylabel('Category')
plt.show()
```

## Questions and Answers

#### What is the average rating for each product category?

```python
if df["rating"].dtype == "object":
    df["rating"] = pd.to_numeric(df["rating"], errors="coerce")

# Calculate the average ratings and sort the results
average_ratings = df.groupby("category")["rating"].mean().reset_index()
average_ratings = average_ratings.sort_values(by="rating", ascending=False)

# Rename the column for clarity
average_ratings.rename(columns={'rating': 'Average Rating'}, inplace=True)

styled_table = average_ratings.head(15).style \
    .background_gradient(cmap='viridis', subset=['Average Rating']) \
    .format({'Average Rating': '{:.2f}'}) \
    .set_caption("Average Product Rating by Category") \
    .set_properties(**{'text-align': 'left'})

styled_table
```

The output shows that most product categories have generally positive customer feedback, with average ratings above 3.50.

### What are the top rating_count products by category?

```python
top_reviewed_per_category = (
    df.groupby("category")
    .apply(lambda x: x.nlargest(10, "rating_count"))
    .reset_index(drop=True)
)

print(top_reviewed_per_category)
```

- The output highlights products likely to be popular within their categories based on high review counts, suggesting customer interest and engagement.
- Review counts range from 9 to 15867, implying varying levels of attention and feedback across products.
- Most listed products have ratings above 3.5, indicating a generally positive customer experience.
- Products with the highest review counts within their categories might be considered potential top sellers, even without direct sales data.

```python
pip install tabulate
```

```python
from tabulate import tabulate

# Sort the DataFrame by 'rating_count' in descending order
most_popular_products = df.sort_values(by='rating_count', ascending=False)

# Select the top 10 products and the columns we want to display
top_10_table = most_popular_products[['product_name', 'rating_count', 'rating']].head(10)

# Rename columns for clarity
top_10_table.rename(columns={
    'product_name': 'Product Name',
    'rating_count': 'Number of Ratings',
    'rating': 'Avg. Rating'
}, inplace=True)

# Print the result in a clean table format
print(tabulate(top_10_table, headers='keys', tablefmt='psql', showindex=False))
```

Based on the information in the table, the most popular product is the Amazon Basics High-Speed HDMI Cable, 6 Feet (2-Pack), Black, with 426,973 ratings.
